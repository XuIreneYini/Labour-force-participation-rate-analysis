```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("D:/25fallterm/ds1/Labour force participation rate analysis/raw_data.csv")
df.head()
```

```python
# Filter data for USA and European Union
usa_data = df[df['REF_AREA'] == 'USA']
eu_data = df[df['REF_AREA'] == 'EU27_2020']

# Sort data by year
usa_data = usa_data.sort_values('TIME_PERIOD')
eu_data = eu_data.sort_values('TIME_PERIOD')
```

```python
# Create trend figure
plt.figure(figsize=(10, 6))

plt.plot(usa_data['TIME_PERIOD'], usa_data['OBS_VALUE'], marker='o', label='United States', linewidth=2)
plt.plot(eu_data['TIME_PERIOD'], eu_data['OBS_VALUE'], marker='s', label='European Union (27)', linewidth=2)

# Set chart titles and labels
plt.title('Labor Force Participation Rate (Ages 15-24)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Participation Rate (%)', fontsize=12)
plt.legend(title='Region')
plt.grid(True, linestyle='--', alpha=0.7)

# Display figure
plt.tight_layout()
plt.show()
```
